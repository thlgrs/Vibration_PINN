# Colab GPU Run: Experiment 01\n\nThis notebook runs `experiments/exp01_clean_data.py` on Google Colab with GPU acceleration.\n\n1. In Colab, set **Runtime -> Change runtime type -> Hardware accelerator = GPU**.\n2. Set `REPO_URL` to your GitHub repository URL.\n3. Run cells in order.

```python
import os
import subprocess
from pathlib import Path

REPO_URL = "https://github.com/YOUR-ORG/Vibration_PINN.git"  # <- set this
PROJECT_DIR = Path("/content/Vibration_PINN")

def can_access_repo(url: str) -> bool:
    result = subprocess.run(
        ["git", "ls-remote", url],
        stdout=subprocess.DEVNULL,
        stderr=subprocess.DEVNULL,
        check=False,
    )
    return result.returncode == 0

if "YOUR-ORG" in REPO_URL:
    raise ValueError("Set REPO_URL to your real GitHub repo first.")

if PROJECT_DIR.exists():
    print(f"Using existing folder: {PROJECT_DIR}")
else:
    print(f"Cloning {REPO_URL} -> {PROJECT_DIR}")
    if not can_access_repo(REPO_URL):
        raise RuntimeError(
            "Cannot access REPO_URL. Check for a typo in owner/repo, verify visibility (public/private), or use auth for private repos."
        )
    subprocess.check_call(["git", "clone", REPO_URL, str(PROJECT_DIR)])

os.chdir(PROJECT_DIR)
print(f"Working directory: {Path.cwd()}")
```
